```python
import pandas as pd
from rake_nltk import Rake
```

```python
rake_nltk_var = Rake()
```

```python
df = pd.read_json('matryoshka/matryoshka.json',
                  orient='index')
df.head()
```

```python
texts = df.text.to_list()
texts[111]
```

```python
result = []
for text in texts:
    sent = ' '.join(text)
    rake_nltk_var.extract_keywords_from_text(sent)
    keyword_extracted = rake_nltk_var.get_ranked_phrases()
    result.append(keyword_extracted)
print(result)
```

```python
df.insert(df.shape[1],
          'key_words',
          result)
df.sample(7)
```

```python
df.to_json('matryoshka_key_words.json',
           orient='records',
           force_ascii=False)
```
